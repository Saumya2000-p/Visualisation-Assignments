```python
Answer1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three-dimensional space.
     np.random.seed(30)
     data = {
          ‘X’: np.random.uniform(-10, 10, 300),
          ‘Y’: np.random.uniform(-10, 10, 300),
          ‘Z’: np.random.uniform(-10, 10, 300),
      }
     df = pd.DataFrame (data)
```

```python
To generate a 3D scatter plot using the provided dataset, we'll need to use libraries like Matplotlib and NumPy. Here’s a step-by-step guide on how to create the plot:

1. Import the necessary libraries.
2. Generate the data.
3. Create the 3D scatter plot.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set the random seed for reproducibility
np.random.seed(30)

# Generate the dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300),
}
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(df['X'], df['Y'], df['Z'], c='blue', marker='o')

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scatter Plot of Random Data Points')

# Show the plot
plt.show()
```

```python
Explanation:-
* Imports: The code imports the necessary libraries.
* Random Data Generation: The dataset is created using NumPy's uniform function.
* 3D Scatter Plot: Matplotlib's scatter method is used to plot the data points in 3D space.
* Labels and Title: The axes are labeled, and a title is added for clarity.
```

```python
To create a heatmap that visualizes the variation in sales across different months and days, we can use Seaborn along with Matplotlib. Here’s how we can do it:

1. Import the necessary libraries.
2. Generate the data.
3. Pivot the data for the heatmap.
4. Create the heatmap.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(20)

# Generate the dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix for the heatmap
heatmap_data = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum', fill_value=0)

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='YlGn', annot=True, fmt='g')

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Day')
plt.title('Heatmap of Sales Variation Across Months and Days')

# Show the plot
plt.show()
```

```python
Explanation:-
1. Imports: The necessary libraries (NumPy, Pandas, Seaborn, and Matplotlib) are imported.
2. Data Generation: A DataFrame is created with random months, days, and sales figures.
3. Pivoting: The data is pivoted to create a matrix where rows represent days, columns represent months, and cell values represent total sales.
4. Heatmap: Seaborn's heatmap function is used to create the visualization. The cmap parameter sets the color palette, and annot=True adds sales figures to the cells.
5. Labels and Title: The axes are labeled, and a title is added for clarity.
```

```python
To create a 3D surface plot for the function 𝑧=sin(√x2+y2) using the provided x and y data, we can use Matplotlib's plot_surface function. Here's how we can do it:

1. Import the necessary libraries.
2. Generate the data.
3. Create the 3D surface plot.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z using the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame (optional, for reference)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
} 
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of z = sin(√(x² + y²))')

# Show the plot
plt.show()
```

```python
Explanation:-
1. Imports: The necessary libraries (NumPy, Pandas, and Matplotlib) are imported.
2. Data Generation: The x and y arrays are created using np.linspace and then meshed together using np.meshgrid. The z values are calculated using the function provided.
3. DataFrame Creation: A DataFrame is created from the flattened arrays (optional if you only need the plot).
4. 3D Surface Plot: The surface is plotted using ax.plot_surface, and a colormap (viridis) is applied for better visualization.
5. Labels and Title: The axes are labeled, and a title is added for clarity.
```

```python
To create a bubble chart that represents each country's population (y-axis), GDP (x-axis), and bubble size proportional to the population, we can use Matplotlib. Here’s how we can do it:

1. Import the necessary libraries.
2. Generate the data.
3. Create the bubble chart.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(25)

# Generate the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], 
            s=df['Population'] * 10,  # Bubble size (scaled for visibility)
            alpha=0.6, 
            edgecolors="w", 
            linewidth=2)

# Annotate each bubble with the country name
for i in range(len(df)):
    plt.annotate(df['Country'][i],
                 (df['GDP'][i], df['Population'][i]),
                 fontsize=10, ha='center')

# Set labels and title
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.title('Bubble Chart of Countries by Population and GDP')

# Show the plot
plt.grid()
plt.show()
```

```python
Explanation:-
1. Imports: The necessary libraries (NumPy, Pandas, and Matplotlib) are imported.
2. Data Generation: A DataFrame is created with random populations and GDP values for a few countries.
3. Bubble Chart: The scatter function is used to create the bubble chart. The size of each bubble is scaled based on the population for better visibility.
4. Annotations: Each bubble is annotated with the corresponding country name for clarity.
5. Labels and Title: The axes are labeled, and a title is added. A grid is included for better readability.
```
