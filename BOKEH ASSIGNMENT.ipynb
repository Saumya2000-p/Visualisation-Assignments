```python
Answer1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x = np.linspace(0, 10, 100)  # x values from 0 to 10
y = np.sin(x)                # y values as the sine of x

# Create a new plot
output_notebook()  # Use this if you want to display it in a Jupyter notebook
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add the sine wave line
p.line(x, y, legend_label='sin(x)', line_width=2, line_color='blue')

# Show the plot
show(p)
```

```python
This code sets up a Bokeh plot of the sine function over the interval from 0 to 10. You can run this in a Jupyter notebook or any Python environment that supports Bokeh. If you're running this outside a notebook, replace output_notebook() with output_file("sine_wave.html") to save the plot as an HTML file.
```

```python
Answer2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

# Prepare the data
n = 100  # number of points
np.random.seed(42)  # for reproducibility

# Generate random data
x = np.random.rand(n) * 10  # x values
y = np.random.rand(n) * 10  # y values
sizes = np.random.randint(10, 100, size=n)  # random sizes for markers
colors = np.random.rand(n)  # random colors

# Create a DataFrame for better handling
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a ColumnDataSource
source = ColumnDataSource(data=data)

# Create a new plot
output_notebook()  # Use this if you want to display it in a Jupyter notebook
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Create a scatter plot with varying sizes and colors
p.scatter('x', 'y', size='sizes', color=linear_cmap('colors', 'Viridis256'), source=source)

# Show the plot
show(p)
```

```python
Explanation:-
1. Data Generation: Random values for x, y, sizes, and colors are generated using NumPy.
2. ColumnDataSource: This is used to efficiently manage the data to be plotted.
3. Plot Creation: A Bokeh figure is created, and a scatter plot is added with the scatter method, using the sizes and colors from the DataFrame.
4. Color Mapping: The linear_cmap function maps the colors to a colormap (in this case, Viridis).
```

```python
Answer3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
     fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
     counts = [20, 25, 30, 35]
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new plot
output_notebook()  # Use this if you want to display it in a Jupyter notebook
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, 
           tools="")

# Add bars to the plot
p.vbar(x='fruits', top='counts', width=0.9, source=source, 
       line_color='white', fill_color='blue')

# Add labels and formatting
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.title.align = "center"

# Show the plot
show(p)
```

```python
Explanation:-
1. Data Preparation: The fruits and counts lists represent the categories and their respective counts.
2. ColumnDataSource: This helps manage the data for the plot.
3. Bar Chart Creation: A vertical bar chart (vbar) is created using the fruits as the x-axis and counts as the heights of the bars.
4. Customization: Basic customization is applied, such as grid line removal, axis labels, and title alignment.
```

```python
Answer4. Create a Bokeh histogram to visualize the distribution of the given data.
     data_hist = np.random.randn(1000)
     hist, edges = np.histogram(data_hist, bins=30)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data
data_hist = np.random.randn(1000)

# Create a histogram
hist, edges = np.histogram(data_hist, bins=30)

# Prepare the data for Bokeh
x = edges[:-1]  # left edges of the bins
width = edges[1:] - edges[:-1]  # width of the bins

# Create a new plot
output_notebook()  # Use this if you want to display it in a Jupyter notebook
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency')

# Add the histogram bars
p.vbar(x=x, top=hist, width=width, fill_color='blue', line_color='white')

# Show the plot
show(p)
```

```python
Explanation:-
1. Data Generation: The data_hist array is filled with 1000 random numbers drawn from a standard normal distribution.
2. Histogram Creation: The np.histogram function computes the histogram data, returning the counts (hist) and the bin edges (edges).
3. Data Preparation: The x-coordinates for the bars are taken from the left edges of the bins, and the width of the bars is calculated.
4. Plot Creation: A Bokeh figure is created, and the vbar method is used to draw the bars of the histogram.
5. Customization: Basic customization, including titles and axis labels, is applied.
```

```python
Answer5. Create a Bokeh heatmap using the provided dataset.
     data_heatmap = np.random.rand(10, 10)
     x = np.linspace(0, 1, 10)
     y = np.linspace(0, 1, 10)
     xx, yy = np.meshgrid(x, y)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.colors import RGB

# Generate random data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Prepare the data for the heatmap
x_flat = np.repeat(x, len(y))
y_flat = np.tile(y, len(x))
z_flat = data_heatmap.flatten()

# Create a color mapper
mapper = linear_cmap(field_name='z', palette='Viridis256', low=z_flat.min(), high=z_flat.max())

# Create a ColumnDataSource
from bokeh.models import ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, z=z_flat))

# Create a new plot
output_notebook()  # Use this if you want to display it in a Jupyter notebook
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           tools="", width=400, height=400)

# Add squares to the plot
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        line_color=None, fill_color=mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```

```python
Explanation:-
1. Data Generation: A 10x10 array of random values is created with data_heatmap.
2. Coordinate Preparation: The x and y arrays are generated, and meshgrid is used to create a grid of coordinates.
3. Flattening Data: The x, y, and z values are flattened for use in the heatmap.
4. Color Mapping: A color mapper is created using linear_cmap to map the values of z to colors from the Viridis palette.
5. Plot Creation: A Bokeh figure is created, and rect is used to add squares for each data point, colored according to the mapped values.
6. Color Bar: A color bar is added to the plot for reference.
```
